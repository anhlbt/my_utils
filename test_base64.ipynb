```python
# coding: utf-8
%matplotlib inline
import matplotlib.pyplot as plt
import base64
import numpy as np
import cv2

img_file = open(r'/home/anhlbt/Pictures/barack-obama-12782369-1-402.jpg', 'rb')  

img_b64encode = base64.b64encode(img_file.read())  # base64

img_file.close()  # 

# print(img_b64encode)

img_b64decode = base64.b64decode(img_b64encode)  # base64

# print(img_b64decode)

img_array = np.frombuffer(img_b64decode, np.uint8)  # 
img = cv2.imdecode(img_array, cv2.COLOR_BGR2RGB)  # 
plt.imshow(img)
```

```python
cv2.imwrite('obama.jpg', img)
```
